```python
target_name = "ARIZABAL" 
SEED_NUM = 7 
redacted_name = ""

for char in target_name:
    if char in "AEIOU":
        redacted_name += str(SEED_NUM) 
    else:
        redacted_name += char 
print(redacted_name)
```

```python
SEED_NUM = 7

engineer_name = "ARIZABAL" 
total_loops = 10

first_letter = engineer_name[0]
ascii_val = ord(first_letter)

verification_hash = (ascii_val * SEED_NUM) - total_loops

print(f"Engineer: {engineer_name}")
print(f"ASCII Value: {ascii_val}")
print(f"Verification Hash: {verification_hash}")
```

```python
correct_password = "BSECE_1A" 
attempts = 0
user_input = ""

while user_input != correct_password:
    user_input = input("Enter Password to Unlock System: ")
    attempts += 1
    if user_input == correct_password:
        print("System Unlocked.")
    else:
        print(f"Access Denied. Attempt {attempts}")

print(f"Correct Password Set: {correct_password}")
print(f"Attempts Before Success: {attempts}")
```
