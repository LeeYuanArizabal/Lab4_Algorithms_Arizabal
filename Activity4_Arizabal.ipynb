```python
# Algorithm: Find the largest number
# 1. Assume first number is the largest
# 2. Compare each number with current largest
# 3. Update if a larger one is found
# 4. Return the largest

numbers = [10, 5, 20, 8, 15]
largest = numbers[0]
for num in numbers:
    if num > largest:
        largest = num
print("Largest number:", largest)

# TRY IT: Find smallest
smallest = numbers[0]
for num in numbers:
    if num < smallest:
        smallest = num
print("Smallest number:", smallest)
```

```python
# O(n) — linear
def print_items(items):
    for item in items:
        print(item)

# O(1) — constant
def print_first(items):
    print(items[0])

# O(n²) — quadratic (nested loop)
def print_pairs(items):
    for i in items:
        for j in items:
            print(i, j)

sample = [1, 2, 3, 4, 5]
print("=== O(n) ===")
print_items(sample)
print("=== O(1) ===")
print_first(sample)
print("=== O(n²) pairs (first 3 only shown) ===")
count = 0
for i in sample:
    for j in sample:
        print(i, j)
        count += 1
        if count >= 3: break
    if count >= 3: break
```

```python
def linear_search(arr, target):
    comparisons = 0
    for i in range(len(arr)):
        comparisons += 1
        if arr[i] == target:
            print(f"Found at index: {i} (after {comparisons} comparisons)")
            return i
    print(f"Not found (made {comparisons} comparisons)")
    return -1

data = [3, 5, 2, 4, 9]
linear_search(data, 4)
linear_search(data, 9)
```

```python
def binary_search(arr, target):
    low, high, comparisons = 0, len(arr) - 1, 0
    while low <= high:
        mid = (low + high) // 2
        comparisons += 1
        if arr[mid] == target:
            print(f"Found at index {mid} after {comparisons} comparison(s)")
            return mid
        elif arr[mid] < target:
            low = mid + 1
        else:
            high = mid - 1
    print(f"Not found after {comparisons} comparison(s)")
    return -1

sorted_data = [1, 3, 5, 7, 9, 11]
binary_search(sorted_data, 7)
binary_search(sorted_data, 1)
binary_search(sorted_data, 11)
```

```python
def bubble_sort(arr):
    n = len(arr)
    swaps = 0
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
                swaps += 1
    print(f"Sorted: {arr}")
    print(f"Total swaps: {swaps}")

to_sort = [64, 34, 25, 12, 22, 11, 90]
bubble_sort(to_sort)
```

```python
def selection_sort(arr):
    comparisons, swaps = 0, 0
    for i in range(len(arr)):
        min_idx = i
        for j in range(i + 1, len(arr)):
            comparisons += 1
            if arr[j] < arr[min_idx]:
                min_idx = j
        arr[i], arr[min_idx] = arr[min_idx], arr[i]
        swaps += 1
    print(f"Sorted: {arr}")
    print(f"Comparisons: {comparisons}, Swaps: {swaps}")

selection_sort([29, 10, 14, 37, 13])
```

```python
def insertion_sort(arr):
    comparisons, swaps = 0, 0
    for i in range(1, len(arr)):
        key = arr[i]
        j = i - 1
        while j >= 0 and key < arr[j]:
            comparisons += 1
            arr[j + 1] = arr[j]
            swaps += 1
            j -= 1
        arr[j + 1] = key
    print(f"Sorted: {arr}")
    print(f"Comparisons: {comparisons}, Swaps: {swaps}")

insertion_sort([12, 11, 13, 5, 6])

# Try with already sorted:
insertion_sort([1, 2, 3, 4, 5])
```
